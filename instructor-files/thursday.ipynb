```python
class Person:
    def greeting(self):
        #print(id(self))
        print('Hello my name is', self.name,f'and I\'m {self.age} years old')
    def get_name(self):
        return self.name.upper()
    def __init__(self,p_name,p_age=0):
        self.name=p_name
        self.age=p_age+1
    def __add__(self,other):
        return self.age+other.age
peter=Person('Peter')
peter.greeting()
jan=Person('Honza',25)
jan.greeting()
print(peter+jan)
```

```python
class Person:
    def greeting(self):
        #print(id(self))
        print('Hello my name is', self.name,f'and I\'m {self.age} years old')
    def get_name(self):
        return self.name.upper()
    def __init__(self,p_name,p_age=0):
        self.name=p_name
        try:
            self.age=int(p_age)
        except:
            raise ValueError('Invalid age')
    def __add__(self,other):
        return self.age+other.age
peter=Person('Peter')
peter.greeting()
jan=Person('Honza',25)
jan.greeting()
print(peter+jan)
x=Person('X','x')
```
