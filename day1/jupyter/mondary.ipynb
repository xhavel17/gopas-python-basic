```python
# builtin types
# numeric types
# Assignment statement, assign value to variable 
salary = 123
year = 2024
print(salary,year)

#separator can be any string
print(salary,year,sep='')

# Language is  case sensitive (identifiers)
# print(Salary) is not print(salary) #MameErroL name "Salary" is not defined 
```

```python
# Integers
salary = 34900

# print the data type use the fce type()
print(salary,type(salary))

# minus value
salary = -100
print(salary)

# hex number
salary=0xff
print(salary)

# binary 
salary = 0b1100
print(salary)

#octal
salary = 0o777 
print(salary)
# starting 3.6 we can use underscore in numeric literals between digits
salary = 100_000_000
print(salary)

# Integers can be large ( unlimited on scale)
# the integer has limit 4300 digits, but can be change ... since python 3.6 ???
x = 123**456
print(x, type(x),x.bit_length())
```
